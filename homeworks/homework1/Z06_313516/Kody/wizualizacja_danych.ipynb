Wizualizacje wyników modeli dla poszczególnych zbiorów

```python
sns.set_theme(style="whitegrid")
```

```python
def plot_scores(rs_df,bs_df, title):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(15, 5)
    sns.scatterplot(x=np.linspace(0, 200, 200), y=rs_df['mean_test_score'],hue=rs_df['mean_fit_time'],ax=ax1).set(title=title + ' - Random Search - Wyniki', xlabel='Numer iteracji', ylabel='Wynik (R2)')
    sns.scatterplot(x=np.linspace(0, 200, 200), y=bs_df['mean_test_score'],hue=bs_df['mean_fit_time'],ax=ax2).set(title=title + ' - Bayesian Search - Wyniki', xlabel='Numer iteracji', ylabel='Wynik (R2)')
    min_ylim = min(ax1.get_ylim()[0], ax2.get_ylim()[0])
    max_ylim = max(ax1.get_ylim()[1], ax2.get_ylim()[1])
    ax1.set_ylim(min_ylim, max_ylim)
    ax2.set_ylim(min_ylim, max_ylim)
    ax1.legend(title = 'Czas uczenia', loc='lower right')
    ax2.legend(title = 'Czas uczenia', loc='lower right')
    plt.show()
```

```python
plot_scores(rs_abalone_rfr,bs_abalone_rfr, 'Abalone - Random Forest')
plot_scores(rs_abalone_gbr,bs_abalone_gbr, 'Abalone - Gradient Boosting')
plot_scores(rs_abalone_en,bs_abalone_en, 'Abalone - Elastic Net')
plot_scores(rs_autompg_rfr,bs_autompg_rfr, 'Auto MPG - Random Forest')
plot_scores(rs_autompg_gbr,bs_autompg_gbr, 'Auto MPG - Gradient Boosting')
plot_scores(rs_autompg_en,bs_autompg_en, 'Auto MPG - Elastic Net')
plot_scores(rs_insurance_rfr,bs_insurance_rfr, 'Insurance - Random Forest')
plot_scores(rs_insurance_gbr,bs_insurance_gbr, 'Insurance - Gradient Boosting')
plot_scores(rs_insurance_en,bs_insurance_en, 'Insurance - Elastic Net')
plot_scores(rs_concrete_rfr,bs_concrete_rfr, 'Concrete - Random Forest')
plot_scores(rs_concrete_gbr,bs_concrete_gbr, 'Concrete - Gradient Boosting')
plot_scores(rs_concrete_en,bs_concrete_en, 'Concrete - Elastic Net')
```

```python
def plot_scores_on_dataset(rs_abalone,rs_autompg,rs_insurance,rs_concrete,bs_abalone,bs_autompg,bs_insurance,bs_concrete, title):
    df = pd.DataFrame({'rs_abalone':rs_abalone['mean_test_score'],'bs_abalone':bs_abalone['mean_test_score'],'rs_autompg':rs_autompg['mean_test_score'],'bs_autompg':bs_autompg['mean_test_score'],'rs_insurance':rs_insurance['mean_test_score'],'bs_insurance':bs_insurance['mean_test_score'],'rs_concrete':rs_concrete['mean_test_score'],'bs_concrete':bs_concrete['mean_test_score']})
    fig = plt.gcf()
    fig.set_size_inches(8, 5)
    sns.violinplot(data=df).set(title=title + ' - rozkład wyników', ylabel='Wynik (R2)')
    plt.xticks(rotation=-45)
    plt.show()
```

```python
plot_scores_on_dataset(rs_abalone_rfr,rs_autompg_rfr,rs_insurance_rfr,rs_concrete_rfr,bs_abalone_rfr,bs_autompg_rfr,bs_insurance_rfr,bs_concrete_rfr, 'Random Forest Regressor')
plot_scores_on_dataset(rs_abalone_gbr,rs_autompg_gbr,rs_insurance_gbr,rs_concrete_gbr,bs_abalone_gbr,bs_autompg_gbr,bs_insurance_gbr,bs_concrete_gbr, 'Gradient Boosting Regressor')
plot_scores_on_dataset(rs_abalone_en,rs_autompg_en,rs_insurance_en,rs_concrete_en,bs_abalone_en,bs_autompg_en,bs_insurance_en,bs_concrete_en, 'Elastic Net')
```

Wizualizacja zbieżności wyników

```python
def plot_current_best_score(rs_df,bs_df, title):
    fig = plt.gcf()
    fig.set_size_inches(8, 5)
    rs_current_best_score = rs_df['mean_test_score'].cummax()
    bs_current_best_score = bs_df['mean_test_score'].cummax()
    current_best_score = pd.DataFrame({'Random Search': rs_current_best_score, 'Bayesian Search': bs_current_best_score})
    sns.lineplot(current_best_score).set(title=title + ' - Zbieżność wyników', xlabel='Numer iteracji', ylabel='Najlepszy wynik (R2)')
    plt.legend(loc='lower right')
    plt.show()
```

```python
plot_current_best_score(rs_abalone_rfr, bs_abalone_rfr, 'Abalone - Random Forest')
plot_current_best_score(rs_abalone_gbr, bs_abalone_gbr, 'Abalone - Gradient Boosting')
plot_current_best_score(rs_abalone_en, bs_abalone_en, 'Abalone - Elastic Net')
plot_current_best_score(rs_autompg_rfr, bs_autompg_rfr, 'Auto MPG - Random Forest')
plot_current_best_score(rs_autompg_gbr, bs_autompg_gbr, 'Auto MPG - Gradient Boosting')
plot_current_best_score(rs_autompg_en, bs_autompg_en, 'Auto MPG - Elastic Net')
plot_current_best_score(rs_insurance_rfr, bs_insurance_rfr, 'Insurance - Random Forest')
plot_current_best_score(rs_insurance_gbr, bs_insurance_gbr, 'Insurance - Gradient Boosting')
plot_current_best_score(rs_insurance_en, bs_insurance_en, 'Insurance - Elastic Net')
plot_current_best_score(rs_concrete_rfr, bs_concrete_rfr, 'Concrete - Random Forest')
plot_current_best_score(rs_concrete_gbr, bs_concrete_gbr, 'Concrete - Gradient Boosting')
plot_current_best_score(rs_concrete_en, bs_concrete_en, 'Concrete - Elastic Net')
```
